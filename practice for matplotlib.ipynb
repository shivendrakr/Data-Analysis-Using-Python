```python
import matplotlib.pyplot as plt 
```

```python
%matplotlib inline
```

```python
import numpy as np
x= np.linspace (0,5,11)
y=x**2
```

```python
x
```

```python
y
```

```python
#functional
plt.plot(x,y)
plt.show() #  it is printing the plot insted of showing the plot 
```

```python
plt.plot(x,y)
plt.xlabel('X Labels')
plt.ylabel('Y labels')
plt.title('Title')
```

```python
# multipl plots
plt.subplot(1,2,1)
plt.plot(x,y,'r')

plt.subplot(1,2,2)
plt.plot(y,x,'b')
```

```python
# Object oriented method 
fig = plt.figure()

axes= fig.add_axes([0.1,0.1,0.8,0.8])

axes.plot(x,y) # same plot using object oriented method
axes.set_xlabel('X labels')
axes.set_ylabel('Y labels')
```

```python
fig= plt.figure()
axes1= fig.add_axes([0.1,0.1,0.8,0.8])
axes2= fig.add_axes([0.5,0.15,0.4,0.3])
```

```python
fig= plt.figure()
axes1= fig.add_axes([0.1,0.1,0.8,0.8])
axes2= fig.add_axes([0.2,0.5,0.4,0.3])

axes1.plot(x,y)
axes1.set_title('Larger plot')

axes2.plot(y,x)
axes2.set_title('Smalller plot')
```

```python
fig=plt.figure()
axes1= fig.add_axes([0.1,0.1,0.8,0.8])
axes1.plot(x,y)
```

```python
# how to create subplot using oo
fig,axes= plt.subplots() # we will get subplot after this 
axes.plot(x,y)
```

```python
fig,axes= plt.subplots(nrows=3,ncols=3)
plt.tight_layout() # it will fix the overlapping problem 
```

```python
fig,axes= plt.subplots(nrows=1,ncols=2)
for current_ax in axes :
    current_ax.plot(x,y)
```

```python
fig,axes= plt.subplots(nrows=1,ncols=2)
axes[0].plot(x,y)
axes[0].set_title('first plot')

axes[1].plot(y,x)
axes[1].set_title('second plot')
plt.tight_layout()
```

## Figure Size and DPI

```python
fig = plt.figure(figsize=(8,2))

ax= fig.add_axes([0,0,1,1])
```

```python
fig,axes = plt.subplots(nrows=2,ncols=1,figsize=(8,2))

axes[0].plot(x,y)

axes[1].plot(y,x)

plt.tight_layout()
```

```python
fig.savefig('my_picture.png',dpi=200) # important line of code 
```

```python
fig=plt.figure()
ax= fig.add_axes([0,0,1,1])
ax.plot(x,y)

ax.set_title('Title')
ax.set_ylabel('Y')
ax.set_xlabel('X')
```

```python
# legends 

fig=plt.figure()
ax= fig.add_axes([0,0,1,1])


ax.plot(x,x**2,label= 'X square ')
ax.plot(x,x**3,label='X cube')
ax.legend(loc=0) # loc can be 1,2,3,4,5,6......
```

```python
fig=plt.figure()
ax= fig.add_axes([0,0,1,1])


ax.plot(x,x**2,label= 'X square ')
ax.plot(x,x**3,label='X cube')
ax.legend(loc=(0.1,0.1))
```

```python
fig.savefig('my_pictures.png',dpi=200)
```
